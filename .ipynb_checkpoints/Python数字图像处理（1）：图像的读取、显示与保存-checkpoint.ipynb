skimage提供了io模块，顾名思义，这个模块是用来图片输入输出操作的，为了方便练习，提供一个data模块，里面嵌套了一些示例图片，可以直接使用。

引用skimage模块可以使用：

```python
from skimage import io
```

### 一、从外部读取图片并显示

**读取单张彩色RGB图片：** 使用 skimage.io.imread(frame) 函数，带一个参数，表示需要读取的文件路径。

**显示图片：** 使用 skimage.io.imshow(arr) 函数，带一个参数，表示需要显示的 arr 数组（读取的图片以 numpy 数组形式计算）

```python
from skimage import io
img=io.imread('images/dog.png')
io.imshow(img)
io.show()   #显示图片，不加的话有错误
```

**读取单张灰度图片：** 使用 skimage.io.imread(fname,as_grey=True) 函数，第一个参数为图片路径，第二个参数为 as_grey , bool 型值，默认为 False 

```python
from skimage import io
img=io.imread('images/dog.png',as_grey=True)
io.imshow(img)
io.show()
```

### 二、程序自带图片

skimage 程序自带了一些示例图片，如果我们不想从外部读取图片，就可以直接使用这些示例图片：

显示这些图片可以使用如下代码，不带任何参数

<img src="images/skimage.png" style="width:650px;height:200px;">

```python
from skimage import io,data
img=data.camera()
io.imshow(img)
io.show()
```

图片名对应的就是函数名，如 camera 图片对应的就是 camera().

这些示例图片存放在 skimage 的安装目录下，路径名称为 data_dir ，我们可以将这个路劲打印出来看看：

```python
from skimage import data_dir
print(data_dir)
```

也就是说，下面两行读取图片的代码效果是一样的：

```python
from skimage import data_dir,data,io
img1=data.camera()   #读取 camera 图片
img2=io.imread(data_dir+'/camera.png')  #读取 camera 图片
```

### 三、保存图片

使用 io 模块的 imsave(fname,arr) 函数来实现。

第一个参数表示保存的路径和名称，第二个参数表示需要保存的数组变量。

```python
from skimage import io,data
img=data.chelsea()
io.imshow(img)
io.imsave('images/cat.jpg',img)
```

保存图片的同时也起到了转换格式的作用。

如果读取时图片格式为 jpg 图片，保存为 png 格式，则将图片从 jpg 图片转化为 png 图片并保存。

### 四、图片信息

如果想知道一些图片的信息，可以在 Spyder 编辑器的右上角显示：（或者在 Pycharm ）

<img src="images/Spyder.png" style="width:450px;height:250px;">

也可以直接以程序方式打印输出

```python
from skimage import io,data
img=data.chelsea()   #小猫图片
io.imshow(img)
io.show()
print(type(img))     #显示类型
print(img.shape)     #显示尺寸
print(img.shape[0])  #显示宽度
print(img.shape[1])  #显示高度
print(img.shape[2])  #显示通道数
print(img.size)      #显示总像素个数
print(img.max())     #最大像素值
print(img.min())     #最小像素值
print(img.mean())    #像素平均值
```
